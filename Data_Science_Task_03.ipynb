```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import re
import nltk
from nltk import word_tokenize
nltk.download('stopwords')
nltk.download('punkt')
```

```python
validation = pd.read_csv('/content/test.csv', header = None)
train = pd.read_csv('/content/train.csv', header = None)
```

```python
validation.columns = ['id', 'tweet']
train.columns = ['id', 'label', 'tweet']
```

```python
validation.head()
```

```python
train.head()
```

```python
val_df = validation.copy()
train_df = train.copy()
```

```python
val_df['lower']= val_df.tweet.str.lower()
val_df['lower']= [str(data) for data in val_df.lower]
val_df['lower']= val_df.lower.apply(lambda x: re.sub('[^A-Za-z0-9]+',' ', x))
```

```python
val_df.head()
```

```python
train_df['lower']= train_df.tweet.str.lower()
train_df['lower']= [str(data) for data in train_df.lower]
train_df['lower']= train_df.lower.apply(lambda x: re.sub('[^A-Za-z0-9]+',' ', x))
```

```python
train_df.head()
```

```python
#spliting
token_tweet = [word_tokenize(str(word)) for word in train_df.lower]
```

```python
token_tweet
```

```python
token_val = [word_tokenize(str(word)) for word in val_df.lower]
```

```python
token_val
```

```python
stopwords_nlkt = nltk.corpus.stopwords
stop_words = stopwords_nlkt.words('english')
print(stop_words)
```

```python
bow_count = CountVectorizer(tokenizer=word_tokenize, stop_words=stop_words, ngram_range=(1,1))
```

```python
reviews_train, reviews_val = train_test_split(train_df, test_size=0.2, random_state=42)
```

```python
x_train = bow_count.fit_transform(reviews_train.lower)
x_val = bow_count.transform(reviews_val.lower)
```

```python
y_train_bow = reviews_train['label']
y_val_bow = reviews_val['label']
```

```python
y_train_bow
```

```python
y_val_bow
```

```python
LogisticRegression_model = LogisticRegression(penalty='l2', solver="liblinear", max_iter=200)
LogisticRegression_model.fit(x_train, y_train_bow)

test_pred = LogisticRegression_model.predict(x_val)
accuracy_score(y_val_bow, test_pred)
```

```python
x_val_bow = bow_count.transform(val_df.lower)
y_val_bow = val_df['lower']
```

```python
y_val_bow
```

```python
val_res = LogisticRegression_model.predict(x_val_bow)
accuracy_score(y_val_bow, val_res)
```
